```python
import pandas_datareader.data as web
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gc
import datetime    
from hurst import compute_Hc, random_walk
```

```python
%%time
start = datetime.date.today() - datetime.timedelta(days=160)
end = datetime.date.today()
companies_names = pd.read_csv('data/constituents_csv.csv')
companies_list = pd.DataFrame(columns=['Company', 'High', 'Low', 'Open', 'Close', 'Volume', 'Adj Close'])
for name in companies_names["Symbol"].values:
    print(name)
    try:
        if name == "BRK.B" or name == "BF.B":
            continue
        temp = web.DataReader(name, 'yahoo', start, end)
        temp["Company"] = name
        temp["Lag_Open"] = temp["Open"].shift(periods=5)
        temp["Diff_Percent"] = (temp["High"] - temp["Lag_Open"]) / temp["Lag_Open"] * 100
        temp.dropna(inplace=True)
        companies_list = pd.concat([companies_list, temp], axis=0)
    except Exception:
        companies_list["Date"] = companies_list.index
        companies_list.reset_index(drop=True)
        companies_list.to_csv("companies_list_" + str(name) + ".csv")
        print(name, companies_names[companies_names["Symbol"].index.values[0]])
        
companies_list["Date"] = companies_list.index
companies_list.reset_index(drop=True)
companies_list.to_csv("companies_list.csv")
companies_list.info()
```

```python
def concat_companies_data():
    companies_list_1 = pd.read_csv("companies_list.csv")
    companies_list_2 = pd.read_csv("companies_list_0_100.csv")
    companies_list = pd.concat([companies_list_2, companies_list_1])
    companies_list.drop(['Unnamed: 0'], axis=1, inplace=True)
    companies_list
```

```python
def compute_hurst(series):
    series = series.apply(np.exp)
    if series.shape[0] > 100:
        H, c, data = compute_Hc(series, kind='change', simplified=True)
    else:
        H = -1
    return H
```

```python
Hurst = companies_list.groupby("Company")["Diff_Percent"].agg(compute_hurst)
Deviation_hurst = abs(0.5 - Hurst) 
#Deviation_hurst.sort_values(ascending=False)
Hurst.head(5)
```

```python
companies_list["Date"] = pd.to_datetime(companies_list["Date"])

companies_list["Candle_length"] = (companies_list["High"] - companies_list["Low"]) / companies_list["High"]
companies_list["Candle_color"] = companies_list["Open"] - companies_list["Close"]
companies_list["Candle_color"] = companies_list["Candle_color"].apply(np.sign)
```

```python
companies_list[companies_list["Candle_color"] ==1]["Candle_length"]
```

```python
%%time
import gc
import warnings

warnings.filterwarnings("ignore")
window = [4, 9, 16, 25]
growth_list = []
for win in window:
    companies_growth = []
    for company in companies_list.Company.unique():
        temp = companies_list[(companies_list["Company"] == company) & (companies_list["Date"] > pd.Timestamp(datetime.date.today() - datetime.timedelta(days=win)))]
        if temp.shape[0] == 0:
            continue
        temp["Close_lag"] = temp["Close"].shift(periods=1)
        temp["OBV_indicator"] = np.sign(temp["Close"].values - temp["Close_lag"].values)
        temp.dropna(inplace=True)
        diff_percent = temp["Diff_Percent"].sum() / temp.shape[0]
        rs = temp[temp["Candle_color"] == 1]["Candle_length"].sum() / temp[temp["Candle_color"] == -1]["Candle_length"].sum()
        rsi = 100 - 100 / (1 + rs)
        obv = (temp["OBV_indicator"] * temp["Volume"]).sum() / 1000000
        avg_volume = temp["Volume"].mean() / 1000000
        companies_growth.append([company, diff_percent, rsi, obv, avg_volume])
        temp = 0
    companies_growth = pd.DataFrame(companies_growth, columns = ["Company_" + str(win), 
                                                                 "AVG_Diff_" + str(win), 
                                                                 "RSI_" + str(win), 
                                                                 "OBV_" + str(win),
                                                                 "AVG_vol_" + str(win)
                                                                ])
    companies_growth = companies_growth.groupby("Company_" + str(win)).agg("sum")
    companies_growth["Hurst_" + str(win)] = Hurst
    companies_growth["AVG_Diff_r_" + str(win)] = companies_growth["AVG_Diff_" + str(win)].apply(np.round)
    growth_list.append(companies_growth)
```

```python
pd.options.display.max_columns = 100
pd.options.display.max_rows = 100
windowed_growth = pd.concat(growth_list, axis=1)
```

```python
def plot_averages(company="ALGN", averages=[5,10,20,50], column="Diff_Percent"):
    plt.rcParams['figure.figsize'] = (20,10)
    cols = ['MA_' + str(i) for i in averages]
    avgs = pd.DataFrame(columns=cols)
    for avg in averages:
        ma = companies_list[companies_list.Company == company].rolling(avg)[column].sum()/avg
        avgs['MA_' + str(avg)] = ma
    avgs.dropna().plot();
    plt.title(company)
```

```python
plot_averages(company="APA")
```
