```python
import pandas_datareader.data as web
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gc
import datetime    
from hurst import compute_Hc, random_walk
```

```python
%%time
company = "INTC"
start = pd.Timestamp("2005-03-31")
end = pd.Timestamp("2020-09-30")

company_hist = pd.DataFrame(columns=['Company', 'High', 'Low', 'Open', 'Close', 'Volume', 'Adj Close'])
temp = web.DataReader(company, 'yahoo', start, end)
temp["Company"] = company
temp.dropna(inplace=True)
company_hist = pd.concat([company_hist, temp], axis=0)
        
company_hist["Date"] = company_hist.index
company_hist.reset_index(drop=True, inplace=True)
company_hist.to_csv("data/company_hist.csv")
company_hist.info()
```
